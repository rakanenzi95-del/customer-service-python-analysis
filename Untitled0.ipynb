```python
df = pd.read_csv("customer_service_python_project_dataset.csv")

df.head()
```

```python
sns.countplot(x="Channel", data=df)

plt.title("Tickets by Channel")
plt.xlabel("Channel")
plt.ylabel("Number of Tickets")

plt.show()
```

```python
df["Agent_Name"].value_counts().plot(kind="bar")

plt.title("Tickets Handled by Each Agent")
plt.xlabel("Agent")
plt.ylabel("Number of Tickets")

plt.show()
```

```python
csat_by_agent = df.groupby("Agent_Name")["CSAT"].mean()

csat_by_agent.plot(kind="bar")

plt.title("Average Customer Satisfaction (CSAT) by Agent")
plt.xlabel("Agent")
plt.ylabel("Average CSAT")

plt.show()
```

```python
sns.countplot(x="Priority", data=df)

plt.title("Ticket Priority Distribution")
plt.xlabel("Priority Level")
plt.ylabel("Number of Tickets")

plt.show()
```
